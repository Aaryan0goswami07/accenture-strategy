```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Segment': ['Gen Z', 'Millennials', 'Boomers'],
        'Market_Size_M': [120, 280, 90],
        'Growth_%': [18, 12, 5]}

df = pd.DataFrame(data)
print("BEST SEGMENT →", df.loc[df['Growth_%'].idxmax(), 'Segment'])

df.plot(x='Segment', y='Market_Size_M', kind='bar', title='UK Market')
plt.show()
```
